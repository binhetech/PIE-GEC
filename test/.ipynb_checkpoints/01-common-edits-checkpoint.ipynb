```python
import pickle
```

# bea-2019

```python
base="bea"
path_multitoken_inserts=open(f"../pickles/{base}/common_multitoken_inserts.p", "rb")
multitoken_inserts = pickle.load(path_multitoken_inserts)
print(f"{len(multitoken_inserts)} multitoken_inserts={multitoken_inserts}")
```

```python
path_inserts=open(f"../pickles/{base}/common_inserts.p", "rb")
inserts = pickle.load(path_inserts)
print(f"{len(inserts)} inserts={inserts}")
```

```python
path_deletes=open(f"../pickles/{base}/common_deletes.p", "rb")
deletes = pickle.load(path_deletes)
print(f"{len(deletes)} deletes={deletes}")
```
